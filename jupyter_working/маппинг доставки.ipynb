```python
import pandasql as ps
```

```python
join_query = '''
    select t1.name as drop_off_name, t1.pickup_point_address as drop_off_address, 
    t1.instruction as drop_off_instruction, t1.lat as drop_off_lat, t1.lng as drop_off_lng,
    client_id, brand_name, pick_up_address, t2.lat as pick_up_lat,
    t2.lng as pick_up_lng
    from 
    (SELECT 
        * from df_drop_off) as t1
        left join
    (select * from df_pick_up) as t2 on t1.key = t2.key 
    '''
df_joined = ps.sqldf(join_query, locals())
```

```python
query_sorted = '''
    select drop_off_name, drop_off_address, drop_off_instruction, drop_off_lat, drop_off_lng, 
    client_id, brand_name, pick_up_address, pick_up_lat, pick_up_lng, distance, 
    row_number() over (partition by brand_name order by distance asc) as rn
    from df_joined
    order by brand_name asc
    '''
df_sorted = ps.sqldf(query_sorted, locals())
```

```python
df_sorted
```

```python
less_than_02 = []
between_02_05 = []
between_05_1 = []
between_1_15 = []
between_15_2 = []
for i,j in df_sorted.iterrows():
    if j['distance'] <= 0.2:
        less_than_02.append(1)
        between_02_05.append(0)
        between_05_1.append(0)
        between_1_15.append(0)
        between_15_2.append(0)
    elif j['distance'] > 0.2 and j['distance'] <= 0.5:
        less_than_02.append(0)
        between_02_05.append(1)
        between_05_1.append(0)
        between_1_15.append(0)
        between_15_2.append(0)
    elif j['distance'] > 0.5 and j['distance'] <= 1:
        less_than_02.append(0)
        between_02_05.append(0)
        between_05_1.append(1)
        between_1_15.append(0)
        between_15_2.append(0)
    elif j['distance'] > 1 and j['distance'] <= 1.5:
        less_than_02.append(0)
        between_02_05.append(0)
        between_05_1.append(0)
        between_1_15.append(1)
        between_15_2.append(0)
    elif j['distance'] > 1.5 and j['distance'] <= 2:
        less_than_02.append(0)
        between_02_05.append(0)
        between_05_1.append(0)
        between_1_15.append(0)
        between_15_2.append(1)
    elif j['distance'] > 2:
        between_15_2.append(0)
        between_1_15.append(0)
        between_05_1.append(0)
        between_02_05.append(0)
        less_than_02.append(0)
```
